```python
# add noie and remove it using filters
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
image = cv2.imread('image.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# -------------------------------
# Add Gaussian Noise
# -------------------------------
mean = 0
sigma = 25
gaussian_noise = np.random.normal(mean, sigma, gray.shape)
gaussian_noisy = gray + gaussian_noise
gaussian_noisy = np.clip(gaussian_noisy, 0, 255).astype(np.uint8)

# -------------------------------
# Add Salt & Pepper Noise
# -------------------------------
sp_noisy = gray.copy()
prob = 0.05

# Salt noise (white pixels)
salt = np.random.rand(*gray.shape) < prob
sp_noisy[salt] = 255

# Pepper noise (black pixels)
pepper = np.random.rand(*gray.shape) < prob
sp_noisy[pepper] = 0

# -------------------------------
# Apply Mean Filter
# -------------------------------
mean_gaussian = cv2.blur(gaussian_noisy, (3, 3))
mean_sp = cv2.blur(sp_noisy, (3, 3))

# -------------------------------
# Apply Gaussian Filter
# -------------------------------
gauss_gaussian = cv2.GaussianBlur(gaussian_noisy, (3, 3), 0)
gauss_sp = cv2.GaussianBlur(sp_noisy, (3, 3), 0)

# -------------------------------
# Display Results
# -------------------------------
titles = [
    "Original",
    "Gaussian Noise",
    "Mean Filter (Gaussian Noise)",
    "Gaussian Filter (Gaussian Noise)",
    "Salt & Pepper Noise",
    "Mean Filter (S&P Noise)",
    "Gaussian Filter (S&P Noise)"
]

images = [
    gray,
    gaussian_noisy,
    mean_gaussian,
    gauss_gaussian,
    sp_noisy,
    mean_sp,
    gauss_sp
]

plt.figure(figsize=(12, 8))

for i in range(len(images)):
    plt.subplot(3, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#histogram equalization
import cv2
import matplotlib.pyplot as plt

# Read image in grayscale
image = cv2.imread('low_contrast.jpg', 0)

# Apply Histogram Equalization
equalized = cv2.equalizeHist(image)

# -------------------------------
# Plot Images
# -------------------------------
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(equalized, cmap='gray')
plt.title("Equalized Image")
plt.axis('off')

# -------------------------------
# Plot Histograms
# -------------------------------
plt.subplot(2, 2, 3)
plt.hist(image.ravel(), bins=256, range=[0,256])
plt.title("Original Histogram")

plt.subplot(2, 2, 4)
plt.hist(equalized.ravel(), bins=256, range=[0,256])
plt.title("Equalized Histogram")

plt.tight_layout()
plt.show()
```

```python
#thresholding l3
import cv2
import matplotlib.pyplot as plt

# Read image in grayscale
image = cv2.imread('image.jpg', 0)

# -------------------------------
# Global Thresholding (Otsu)
# -------------------------------
ret, otsu_thresh = cv2.threshold(
    image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

# -------------------------------
# Adaptive Thresholding
# -------------------------------
adaptive_mean = cv2.adaptiveThreshold(
    image, 255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY,
    11, 2
)

adaptive_gaussian = cv2.adaptiveThreshold(
    image, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11, 2
)

# -------------------------------
# Display Results
# -------------------------------
titles = [
    "Original Image",
    "Otsu Thresholding",
    "Adaptive Mean",
    "Adaptive Gaussian"
]

images = [
    image,
    otsu_thresh,
    adaptive_mean,
    adaptive_gaussian
]

plt.figure(figsize=(10, 6))

for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#l4 edge detection
import cv2 
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('image.jpg', 0)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel = cv2.magnitude(sobel_x, sobel_y)
sobel = np.uint8(sobel)
kernelx = np.array([[ -1, 0, 1],[ -1, 0, 1],[ -1, 0, 1]])
kernely = np.array([[ 1,  1,  1],[ 0,  0,  0],[-1, -1, -1]])
prewitt_x = cv2.filter2D(image, -1, kernelx)
prewitt_y = cv2.filter2D(image, -1, kernely)
prewitt = cv2.magnitude(np.float32(prewitt_x), np.float32(prewitt_y))
prewitt = np.uint8(prewitt)
canny = cv2.Canny(image, 100, 200) 
titles = ["Original", "Sobel", "Prewitt", "Canny"]
images = [image, sobel, prewitt, canny]  
plt.figure(figsize=(10, 6))
for i in range(4):
    plt.subplot(2, 2, i+1)   
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])  
    plt.axis('off')
plt.tight_layout() 
plt.show()
```

```python
#l5 character segmentation
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('text_image.png')
orig = img.copy()
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)
kernel = np.ones((3,3), np.uint8)
thresh = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
for cnt in contours:
	x, y, w, h = cv2.boundingRect(cnt)
	if w > 10 and h > 10:
		cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 3)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(1,2,2)
plt.title("Character Segmented Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
#rice-l6
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# Step 1: Read Image
# -------------------------------
img = cv2.imread('rice.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# -------------------------------
# Step 2: Thresholding
# -------------------------------
_, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Invert if needed (rice should be white)
thresh = cv2.bitwise_not(thresh)

# -------------------------------
# Step 3: Noise Removal
# -------------------------------
kernel = np.ones((3,3), np.uint8)
clean = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

# -------------------------------
# Step 4: Find Contours
# -------------------------------
contours, _ = cv2.findContours(clean, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# -------------------------------
# Step 5: Measurements
# -------------------------------
count = 0
areas = []
filtered_count = 0

output = img.copy()

for cnt in contours:
    area = cv2.contourArea(cnt)

    # Filter small noise
    if area > 50:
        count += 1
        areas.append(area)

        # Perimeter
        perimeter = cv2.arcLength(cnt, True)

        # Major Axis Length (Ellipse)
        if len(cnt) >= 5:
            ellipse = cv2.fitEllipse(cnt)
            (center, axes, orientation) = ellipse
            major_axis = max(axes)
        else:
            major_axis = 0

        # Area range filter (example: 100–1000)
        if 100 < area < 1000:
            filtered_count += 1

        # Draw contour
        cv2.drawContours(output, [cnt], -1, (0,255,0), 2)

# -------------------------------
# Step 6: Display Results
# -------------------------------
print("Total Rice Count:", count)
print("Rice Count (Area 100–1000):", filtered_count)
print("Average Area:", np.mean(areas))

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Binary Image")
plt.imshow(clean, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Detected Rice Grains")
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

```python
#l7 masking
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# Step 1: Read Image
# -------------------------------
img = cv2.imread('l3.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# -------------------------------
# Step 2: Define 3x3 Mask (Kernel)
# Example: Edge Detection
# -------------------------------
kernel = np.array([[ -1, -1, -1],
                   [ -1,  8, -1],
                   [ -1, -1, -1]])

# -------------------------------
# Step 3: Apply Convolution
# -------------------------------
convolved = cv2.filter2D(gray, -1, kernel)

# -------------------------------
# Step 4: Display Results
# -------------------------------
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("After Convolution (3x3 Mask)")
plt.imshow(convolved, cmap='gray')
plt.axis('off')

plt.show()
```

```python
#l8 decomposition
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pywt

# -------------------------------
# Step 1: Read Image
# -------------------------------
img = cv2.imread('l3.jpg', 0)  # grayscale

# -------------------------------
# Step 2: First Level Decomposition
# -------------------------------
coeffs1 = pywt.dwt2(img, 'haar')
LL1, (LH1, HL1, HH1) = coeffs1

# -------------------------------
# Step 3: Second Level Decomposition
# (Apply again on LL1)
# -------------------------------
coeffs2 = pywt.dwt2(LL1, 'haar')
LL2, (LH2, HL2, HH2) = coeffs2

# -------------------------------
# Step 4: Plot Results
# -------------------------------
plt.figure(figsize=(12,8))

# Level 1
plt.subplot(3,4,1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(3,4,2)
plt.title("LL1")
plt.imshow(LL1, cmap='gray')
plt.axis('off')

plt.subplot(3,4,3)
plt.title("LH1")
plt.imshow(LH1, cmap='gray')
plt.axis('off')

plt.subplot(3,4,4)
plt.title("HL1")
plt.imshow(HL1, cmap='gray')
plt.axis('off')

plt.subplot(3,4,5)
plt.title("HH1")
plt.imshow(HH1, cmap='gray')
plt.axis('off')

# Level 2
plt.subplot(3,4,6)
plt.title("LL2")
plt.imshow(LL2, cmap='gray')
plt.axis('off')

plt.subplot(3,4,7)
plt.title("LH2")
plt.imshow(LH2, cmap='gray')
plt.axis('off')

plt.subplot(3,4,8)
plt.title("HL2")
plt.imshow(HL2, cmap='gray')
plt.axis('off')

plt.subplot(3,4,9)
plt.title("HH2")
plt.imshow(HH2, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
